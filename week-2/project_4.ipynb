```python
def calculate_atr(years_experience, age):
    if years_experience > 25 and age >= 55:
        return 5600000
    elif years_experience > 20 and age >= 45:
        return 4480000
    elif years_experience > 10 and age >= 35:
        return 1500000
    else:
        return 550000

# Example usage
years = float(input("Enter years of experience: "))
age = float(input("Enter age: "))
tax = calculate_atr(years, age)
print(f"Annual Tax Revenue: N{tax:,}")
```
