# QUADRATIC EQUATION

```python
a = int(input("enter a"))
b = int(input("enter b"))
c = int(input("enter c"))

D = (b ** 2) - (4 * a * c)

x1 = ((-b) + (D ** 0.5)) / (2 * a)
x2 = ((-b) - (D ** 0.5)) / (2 * a)

print (f"the roots are {x1} and {x2}")
```

# CUBIC EQUATION

```python
import numpy as np

def solve_cubic(a, b, c, d):
    """Find roots of cubic equation ax³ + bx² + cx + d = 0"""
    coefficients = [a, b, c, d]
    roots = np.roots(coefficients)
    return roots

# Example usage
print("Cubic Equation Solver: ax³ + bx² + cx + d = 0")
print("-" * 40)

try:
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))
    d = float(input("Enter coefficient d: "))
    
    roots = solve_cubic(a, b, c, d)
    
    print("\nRoots:")
    for i, root in enumerate(roots, 1):
        # Check if root is real (imaginary part close to zero)
        if abs(root.imag) < 1e-10:
            print(f"Root {i}: {root.real:.6f}")
        else:
            print(f"Root {i}: {root:.6f}")
            
except ValueError:
    print("Please enter valid numbers!")
except Exception as e:
    print(f"Error: {e}")
```

# QUARTIC

```python
import numpy as np

def solve_quartic(a, b, c, d, e):
    """Find roots of quartic equation ax⁴ + bx³ + cx² + dx + e = 0"""
    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)
    return roots

# Example usage
print("=" * 50)
print("QUARTIC EQUATION SOLVER")
print("ax⁴ + bx³ + cx² + dx + e = 0")
print("=" * 50)

try:
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))
    d = float(input("Enter coefficient d: "))
    e = float(input("Enter coefficient e: "))
    
    roots = solve_quartic(a, b, c, d, e)
    
    print("\nRoots:")
    for i, root in enumerate(roots, 1):
        # Check if root is real (imaginary part close to zero)
        if abs(root.imag) < 1e-10:
            print(f"Root {i}: {root.real:.6f}")
        else:
            print(f"Root {i}: {root:.6f}")
            
except ValueError:
    print("Please enter valid numbers!")
except Exception as e:
    print(f"Error: {e}")
```
