# STORE CHECKOUT

```python
# BEGIN StoreCheckout
BEGIN Checkout
    SET total = 0
REPEAT
INPUT price
IF price > 0 THEN
    total = total + price
END IF
UNTIL price = 0
PRINT "Total: $" + total
INPUT cash
IF cash >= total THEN
        change = cash - total
        PRINT "Change: $" + change
        PRINT "Thank you!"
    ELSE
        PRINT "Not enough money"
    END IF
END
```
