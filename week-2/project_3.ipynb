```python
import pandas as pd
from tabulate import tabulate

# Girls data
girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", 
               "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
girls_ages = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
girls_heights = [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
girls_scores = [80, 85, 70, 60, 76, 68, 87, 95, 50, 49]

# Boys data
boys_names = ["Chinedu", "Emeka", "Oluwaseun", "Michael", "Daniel",
              "John", "Peter", "James", "Joseph", "David"]
boys_ages = [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
boys_heights = [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
boys_scores = [74, 87, 75, 68, 86, 78, 87, 98, 54, 60]

# Combine data
all_names = girls_names + boys_names
all_ages = girls_ages + boys_ages
all_heights = girls_heights + boys_heights
all_scores = girls_scores + boys_scores
genders = ["Girl"] * 10 + ["Boy"] * 10

# Create DataFrame
df = pd.DataFrame({
    "Name": all_names,
    "Gender": genders,
    "Age": all_ages,
    "Height (ft)": all_heights,
    "Score": all_scores
})

print(tabulate(df, headers='keys', tablefmt='grid', showindex=False))
```
