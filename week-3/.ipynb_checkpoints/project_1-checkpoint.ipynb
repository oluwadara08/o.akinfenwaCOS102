```python
import tkinter as tk
import pandas as pd

EMPLOYEES = []

def read_employees(excel_file="GIG-logistics.xlsx"):
    """Read employee data from Excel file. Returns list of employee dictionaries."""
    employees = []
    
    try:
        df = pd.read_excel(excel_file, sheet_name=0)
        
        if 'Name' not in df.columns or 'Department' not in df.columns:
            print(f"ERROR: Excel file must have 'Name' and 'Department' columns. Found: {', '.join(df.columns)}")
            return employees
        
        # Extract name and department
        for _, row in df.iterrows():
            name = str(row['Name']).strip() if pd.notna(row['Name']) else ""
            dept = str(row['Department']).strip() if pd.notna(row['Department']) else ""
            
            if name and dept and name != 'nan' and dept != 'nan':
                employees.append({
                    "name": name,
                    "department": dept,
                    "employee_id": str(row['EmployeeID']) if 'EmployeeID' in df.columns and pd.notna(row['EmployeeID']) else "",
                    "role": str(row['Role']) if 'Role' in df.columns and pd.notna(row['Role']) else "",
                    "email": str(row['Email']) if 'Email' in df.columns and pd.notna(row['Email']) else "",
                    "phone": str(row['Phone']) if 'Phone' in df.columns and pd.notna(row['Phone']) else "",
                    "location": str(row['Location']) if 'Location' in df.columns and pd.notna(row['Location']) else ""
                })
        
        return employees
    
    except FileNotFoundError:
        print(f"ERROR: Excel file '{excel_file}' not found. Please make sure the file is in the same folder.")
        return employees
    except Exception as e:
        print(f"ERROR: Could not read Excel file: {str(e)}")
        return employees

def find_employee(name, department):
    """Check if an employee with given name and department exists.
       Returns employee dict or None."""
    name_clean = name.strip().lower()
    dept_clean = department.strip().lower()
    
    for emp in EMPLOYEES:
        if emp["name"].lower() == name_clean and emp["department"].lower() == dept_clean:
            return emp
    return None

def get_department_colleagues(department, exclude_name):
    """Return list of colleague names (excluding exclude_name) in the given department."""
    dept_clean = department.strip().lower()
    exclude_clean = exclude_name.strip().lower()
    
    colleagues = []
    for emp in EMPLOYEES:
        if emp["department"].lower() == dept_clean and emp["name"].lower() != exclude_clean:
            colleagues.append(emp["name"])
    return colleagues

def get_department_summary(department):
    """Get count and list of all employees in a department."""
    dept_clean = department.strip().lower()
    dept_employees = [emp for emp in EMPLOYEES if emp["department"].lower() == dept_clean]
    return len(dept_employees), [emp["name"] for emp in dept_employees]

def on_verify():
    """Callback for the Verify button."""
    name = entry_name.get().strip()
    department = entry_dept.get().strip()
    
    output_text.config(state=tk.NORMAL)
    output_text.delete(1.0, tk.END)
    
    if not name or not department:
        output_text.insert(tk.END, "⚠️ Please enter both name and department.")
        output_text.config(state=tk.DISABLED)
        return
    
    employee = find_employee(name, department)
    
    if employee:
        # Employee exists - prepare welcome message
        colleagues = get_department_colleagues(department, name)
        total_count, all_names = get_department_summary(department)
        
        # Create formatted welcome message
        welcome_msg = f"✅ WELCOME, {employee['name'].upper()}!\n"
        welcome_msg += f"📋 Department: {employee['department']}\n"
        
        # Add optional details if available
        if employee.get('role'):
            welcome_msg += f"💼 Role: {employee['role']}\n"
        if employee.get('location'):
            welcome_msg += f"📍 Location: {employee['location']}\n"
        
        welcome_msg += f"\n{'='*50}\n"
        welcome_msg += f"📊 DEPARTMENT: {employee['department']}\n"
        welcome_msg += f"👥 Total Members: {total_count}\n"
        
        if colleagues:
            welcome_msg += f"\n👨‍💼 OTHER MEMBERS:\n"
            for i, col in enumerate(colleagues, 1):
                # Try to add role if available
                col_emp = next((e for e in EMPLOYEES if e["name"] == col), None)
                role_info = f" - {col_emp['role']}" if col_emp and col_emp.get('role') else ""
                welcome_msg += f"  {i}. {col}{role_info}\n"
        else:
            welcome_msg += f"\n🌟 You are the only member in this department!\n"
        
        output_text.insert(tk.END, welcome_msg)
        
    else:
        # Employee not found
        output_text.insert(tk.END, f"❌ Sorry, employee '{name}' in department '{department}' does not exist.\n\nPlease check your spelling or contact HR.")
    
    output_text.config(state=tk.DISABLED)

def clear_fields():
    """Clear input fields and output."""
    entry_name.delete(0, tk.END)
    entry_dept.delete(0, tk.END)
    output_text.config(state=tk.NORMAL)
    output_text.delete(1.0, tk.END)
    output_text.config(state=tk.DISABLED)
    entry_name.focus()

def show_all_departments():
    """Show all departments in a new window."""
    dept_window = tk.Toplevel(root)
    dept_window.title("All Departments")
    dept_window.geometry("400x500")
    dept_window.transient(root)
    dept_window.grab_set()
    
    # Count employees per department
    dept_counts = {}
    for emp in EMPLOYEES:
        dept = emp["department"]
        dept_counts[dept] = dept_counts.get(dept, 0) + 1
    
    # Display departments
    tk.Label(dept_window, text="📋 AVAILABLE DEPARTMENTS", 
             font=("Arial", 12, "bold")).pack(pady=10)
    
    dept_listbox = tk.Listbox(dept_window, width=40, height=20)
    dept_listbox.pack(padx=20, pady=10, fill=tk.BOTH, expand=True)
    
    for dept, count in sorted(dept_counts.items()):
        dept_listbox.insert(tk.END, f"{dept} ({count} employees)")
    
    tk.Button(dept_window, text="Close", command=dept_window.destroy,
              bg="#f44336", fg="white", padx=20, pady=5).pack(pady=10)

# --- GUI Setup ---
root = tk.Tk()
root.title("🏢 GIG Logistics - Employee Verification System")
root.geometry("650x550")
root.resizable(False, False)

# Load employee data
EMPLOYEES = read_employees()

if not EMPLOYEES:
    print(f"⚠️ Warning: No employee data loaded. Total employees: {len(EMPLOYEES)}")
else:
    print(f"✅ Loaded {len(EMPLOYEES)} employees")

# Header Frame
header_frame = tk.Frame(root, bg="#2c3e50", height=80)
header_frame.pack(fill=tk.X)
tk.Label(header_frame, text="GIG LOGISTICS", 
         font=("Arial", 20, "bold"), bg="#2c3e50", fg="white").pack(pady=20)

# Input Frame
frame_input = tk.Frame(root, padx=30, pady=20)
frame_input.pack(fill=tk.X)

tk.Label(frame_input, text="Employee Name:", font=("Arial", 10, "bold")).grid(row=0, column=0, sticky=tk.W, pady=10)
entry_name = tk.Entry(frame_input, width=45, font=("Arial", 10))
entry_name.grid(row=0, column=1, pady=10, padx=10)
entry_name.focus()

tk.Label(frame_input, text="Department:", font=("Arial", 10, "bold")).grid(row=1, column=0, sticky=tk.W, pady=10)
entry_dept = tk.Entry(frame_input, width=45, font=("Arial", 10))
entry_dept.grid(row=1, column=1, pady=10, padx=10)

# Button Frame
button_frame = tk.Frame(frame_input)
button_frame.grid(row=2, column=0, columnspan=2, pady=20)

btn_verify = tk.Button(button_frame, text="✓ Verify & Welcome", command=on_verify, 
                       bg="#27ae60", fg="white", padx=20, pady=8, font=("Arial", 10, "bold"))
btn_verify.pack(side=tk.LEFT, padx=5)

btn_clear = tk.Button(button_frame, text="🗑 Clear", command=clear_fields, 
                      bg="#f39c12", fg="white", padx=20, pady=8, font=("Arial", 10, "bold"))
btn_clear.pack(side=tk.LEFT, padx=5)

btn_depts = tk.Button(button_frame, text="📋 View All Departments", command=show_all_departments, 
                      bg="#3498db", fg="white", padx=20, pady=8, font=("Arial", 10, "bold"))
btn_depts.pack(side=tk.LEFT, padx=5)

# Output area
output_frame = tk.Frame(root, padx=30, pady=10)
output_frame.pack(fill=tk.BOTH, expand=True)

tk.Label(output_frame, text="VERIFICATION RESULT:", font=("Arial", 11, "bold"), 
         fg="#2c3e50").pack(anchor=tk.W)

output_text = tk.Text(output_frame, height=18, wrap=tk.WORD, state=tk.DISABLED, 
                      bg="#f8f9fa", relief=tk.SUNKEN, borderwidth=2, font=("Courier", 9))
output_text.pack(fill=tk.BOTH, expand=True, pady=10)

# Status bar
status_frame = tk.Frame(root, bg="#ecf0f1", height=30)
status_frame.pack(fill=tk.X, side=tk.BOTTOM)
status_label = tk.Label(status_frame, text=f"✅ System Ready | {len(EMPLOYEES)} employees loaded", 
                        bg="#ecf0f1", fg="#7f8c8d", font=("Arial", 8))
status_label.pack(side=tk.LEFT, padx=10, pady=5)

# Bind Enter key to verify
root.bind('<Return>', lambda event: on_verify())

# Start the GUI event loop
root.mainloop()
```
