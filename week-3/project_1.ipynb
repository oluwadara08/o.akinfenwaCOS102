```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load dat
df = pd.read_excel("GIG-logistics.xlsx")

def verify_employee():
    name = name_entry.get()
    department = dept_entry.get()

    employee = df[
        (df['Name'].str.lower() == name.lower()) &
        (df['Department'].str.lower() == department.lower())
    ]

    if not employee.empty:
        members = df[
            df['Department'].str.lower() == department.lower()
        ]['Name'].tolist()

        result = f"Welcome {name}!\n\nOther members:\n"

        for member in members:
            if member.lower() != name.lower():
                result += f"- {member}\n"

        output_label.config(text=result)

    else:
        output_label.config(
            text="Employee does not exist."
        )

# GUI
root = tk.Tk()
root.title("GIG Logistics Verification System")
root.geometry("500x400")

tk.Label(root, text="Employee Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(
    root,
    text="Verify",
    command=verify_employee
).pack(pady=10)

output_label = tk.Label(root, text="", justify="left")
output_label.pack()

root.mainloop()
```
